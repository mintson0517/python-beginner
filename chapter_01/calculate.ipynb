```python
#四則演算
a = 10
b = 3
print("a =", a)
print("b =", b)
print("足し算:", a + b)  # 足し算
print("引き算:", a - b)  # 引き算
print("掛け算:", a * b)  # 掛け算
print("割り算:", a / b)  # 割り算
print("整数除算:", a // b) # 整数除算
print("剰余:", a % b)  # 剰余
print("べき乗:", a ** b) # べき乗
```

```python
a = 10
b = 3
a += b
print(a)  # 13
a -= b
print(a)  # 11
a *= b
print(a)  # 44
a /= b
print(a)  # 22.0
a %= b
print(a)  # 1.0
a **= b
print(a)  # 1.0
a //= b
print(a)  # 0.0
```
